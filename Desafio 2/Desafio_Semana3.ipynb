Desafio 2 - Conhecendo melhor nossa base de consumidores: qual estado possui os clientes com melhores pontuações de crédito?

```python
#instalando bibliotecas 
import pandas as pd
import numpy as np
import seaborn as sns
import json
```

```python
#lendo arquivo 
df = pd.read_csv('desafio1.csv')
df.head()
```

```python
#Colunas do Dataframe
df.columns
```

```python
#Dimensão do DataFrame
df.shape
```

```python
#Quantidade de UF com residencia no DataFrame
df['estado_residencia'].value_counts()
```

```python
#Verificar se tem correlação
sns.scatterplot(x='saldo_conta', y='pontuacao_credito', data=df)
```

```python
df[['saldo_conta','pontuacao_credito']].corr()
```

```python
#Conhecendo os dados com Estatistica describe
df.describe()
```

```python
#Agrupamento da pontuação crédito e os UF
#Ultilizando o metodo agg para agrupamento, em seguida renomear as colunas para saída ser visualizado no JSON e finalizando colocar os valores em ordem crescente

df_estado = df.groupby('estado_residencia')['pontuacao_credito'].agg([pd.Series.mode,
                        np.median,
                        np.mean,
                        np.std]).rename(columns = {'mode' : 'moda',
                            'median' : 'mediana',
                            'mean' : 'media',
                            'std' : 'desvio_padrao'}).sort_values(by='estado_residencia', ascending=False)
```

```python
#Transposição DataFrame

df_submission = df_estado.T
df_submission.head()
```

```python
#Transformando o DF em arquivo JSON conforme solicitado no desafio 
df_submission.to_json('submission.json',orient='columns')
```
